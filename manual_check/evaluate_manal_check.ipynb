```python
#import data
data = pd.read_csv('/content/drive/MyDrive/free lance/post classification/manual_check/prediction-manual-check.csv')
data.head(5)
```

```python
bert_pred = pd.read_csv('/content/drive/MyDrive/free lance/post classification transformers/prediction/bert_model_prediction.csv')
bert_pred.head()
```

```python
roberta_pred = pd.read_csv('/content/drive/MyDrive/free lance/post classification transformers/prediction/roberta_model_prediction.csv')
roberta_pred.head()
```

```python
#merge the three dataframe
df1 = pd.merge(data, bert_pred, on='post', how='inner')
final_df = pd.merge(df1, roberta_pred, on='post', how='inner')
```

```python
def modify_string(cell_value):
    if isinstance(cell_value, str):
        return cell_value.strip().lower()
    else:
        return cell_value

final_df['Manual '] = final_df['Manual '].apply(modify_string)
data_checked = final_df[final_df['Manual '] == "checked"]
data_checked.reset_index(inplace=True)
```

```python
# Create a dictionary to map old column names to new names
column_mapping = {'compulsion_x': 'compulsion', 'obs-com_x': 'obs-com','obsession_x':'obsession',
                  'compulsion_y': 'compulsion_bert', 'obs-com_y': 'obs-com_bert','obsession_y':'obsession_bert',
                  'compulsion_z': 'compulsion_roberta', 'obs-com_z': 'obs-com_roberta','obsession_z':'obsession_roberta'}

# Rename the columns
data_checked = data_checked.rename(columns=column_mapping)
data_checked.head(3)
```

```python
data_checked.to_csv('data_checked.csv',index=False)
```

```python
data_checked = pd.read_csv('/content/data_checked.csv')
```

```python
data_checked.columns
```

```python
#extract true label and predicted label
def extract_prediction_values(row):
    return [int(row[col].split('-')[0].strip()) for col in ['compulsion', 'obs-com', 'obsession']]

def extract_prediction_bert_values(row):
    return [int(row[col])  for col in ['compulsion_bert', 'obs-com_bert', 'obsession_bert']]

def extract_prediction_roberta_values(row):
    return [int(row[col])  for col in ['compulsion.1', 'obs-com.1', 'obsession.1']]

def extract_true_values(row):
    return [int(row[col].split('-')[1].strip()) for col in ['compulsion', 'obs-com', 'obsession']]

data_checked['prediction_values'] = data_checked.apply(extract_prediction_values, axis=1)
data_checked['prediction_bert_values'] = data_checked.apply(extract_prediction_bert_values, axis=1)
data_checked['prediction_roberta_values'] = data_checked.apply(extract_prediction_roberta_values, axis=1)
data_checked['true_values'] = data_checked.apply(extract_true_values, axis=1)
```

# Evaluation

### **Machine learning model**

```python
# Generate a classification report
from sklearn.metrics import classification_report

true_labels = data_checked['true_values'].tolist()
predicted_labels = data_checked['prediction_values'].tolist()

report = classification_report(true_labels, predicted_labels, target_names=['compulsion', 'obs-com', 'obsession'])

# Print the classification report
print(report)
```

### **BERT model**

```python
# Generate a classification report
from sklearn.metrics import classification_report

true_labels = data_checked['true_values'].tolist()
predicted_labels = data_checked['prediction_bert_values'].tolist()

report = classification_report(true_labels, predicted_labels, target_names=['compulsion', 'obs-com', 'obsession'])

# Print the classification report
print(report)
```

### **ROBERTA model**

```python
# Generate a classification report
from sklearn.metrics import classification_report

true_labels = data_checked['true_values'].tolist()
predicted_labels = data_checked['prediction_roberta_values'].tolist()

report = classification_report(true_labels, predicted_labels, target_names=['compulsion', 'obs-com', 'obsession'])

# Print the classification report
print(report)
```
