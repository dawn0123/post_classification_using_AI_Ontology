```python
import pandas as pd
```

```python
df_ml = pd.read_csv('/content/drive/MyDrive/post classification/input_data/make_prediction_for_unlabed_data.csv')
df_bert = pd.read_csv('/content/drive/MyDrive/post classification/input_data/bert_model_prediction.csv')
df_manual_check = pd.read_csv('/content/drive/MyDrive/post classification/input_data/prediction-manual-check.csv')
```

```python
# Merge the two datasets based on the common columns
merged_df = df_ml.merge(df_bert, on=['post', 'prefLabel'], suffixes=('_ml', '_bert'))

# Create new columns for combined predictions
merged_df['compulsion_combined'] = merged_df['compulsion_ml'] + merged_df['compulsion_bert']
merged_df['obs-com_combined'] = merged_df['obs-com_ml'] + merged_df['obs-com_bert']
merged_df['obsession_combined'] = merged_df['obsession_ml'] + merged_df['obsession_bert']
```

```python
merged_df
```

```python
# Extract the second value from 'obsession', 'obs-com', and 'compulsion' in the 'other_dataset'
df_manual_check['obsession_first_checker'] = df_manual_check['obsession'].str.split('-').str[1]
df_manual_check['obs-com_first_checker'] = df_manual_check['obs-com'].str.split('-').str[1]
df_manual_check['compulsion_first_checker'] = df_manual_check['compulsion'].str.split('-').str[1]

# Merge the 'second_checker' values to 'merged_df' using the index
merged_df['obsession_first_checker'] = df_manual_check['obsession_first_checker']
merged_df['obs-com_first_checker'] = df_manual_check['obs-com_first_checker']
merged_df['compulsion_first_checker'] = df_manual_check['compulsion_first_checker']
```

```python
merged_df
```

```python
columns_to_remove = ['obsession_with_reasoning', 'obsession_without_reasoning', 'compulsion_combined', 'obs-com_combined', 'obsession_combined']
merged_df.drop(columns=columns_to_remove, inplace=True)
```

```python
merged_df
```

```python

# Save the modified DataFrame to a CSV file
merged_df.to_csv('Post_ModelPredictions_ManualChecking.csv', index=False)
```

```python
# Count the number of 1s for each label
compulsion_count_1 = (merged_df['compulsion_ml'] == 1).sum()
obs_com_count_1 = (merged_df['obs-com_ml'] == 1).sum()
obsession_count_1 = (merged_df['obsession_ml'] == 1).sum()

# Print the counts for each label
print("Number of 1s in Compulsion Label:", compulsion_count_1)
print("Number of 1s in Obs-Com Label:", obs_com_count_1)
print("Number of 1s in Obsession Label:", obsession_count_1)
```

```python
# Count the number of posts where all labels are 0
count_all_labels_0 = ((merged_df['compulsion_ml'] == 0) & (merged_df['obs-com_ml'] == 0) & (merged_df['obsession_ml'] == 0)).sum()

# Print the count
print("Number of posts with all labels equal to 0:", count_all_labels_0)
```

```python
def save_posts_with_all_labels_0(merged_df, output_filename):
    # Select the posts where all labels are equal to 0
    posts_with_all_labels_0 = merged_df[
        (merged_df['compulsion_ml'] == 0) &
        (merged_df['obs-com_ml'] == 0) &
        (merged_df['obsession_ml'] == 0)
    ]

    # Save the selected posts to a CSV file
    posts_with_all_labels_0.to_csv(output_filename, index=False)

# Specify the output filename
output_filename = 'posts_with_all_labels_0.csv'

# Call the function to save the posts with all labels equal to 0
save_posts_with_all_labels_0(merged_df, output_filename)

print(f"Posts with all labels equal to 0 have been saved to {output_filename}")
```
