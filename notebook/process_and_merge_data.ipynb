# read input data

```python
final_annotations = pd.read_csv('../content/drive/My Drive/post classification/input_data/FinalAnnotations_Obs_Com.csv')
compulsion = pd.read_csv('../content/drive/My Drive/post classification/input_data/compulsion-based-reasoning.csv')
obs_com = pd.read_csv('../content/drive/My Drive/post classification/input_data/obs-com-byReasoning.csv')
obession_1 = pd.read_csv('../content/drive/My Drive/post classification/input_data/obsession-based-annotation-reasoning.csv')
obession_2 = pd.read_csv('../content/drive/My Drive/post classification/input_data/obsession-based-annotation.csv')
obession = pd.concat([obession_1,obession_2])
obession = obession.drop_duplicates()
```

# select required columns

```python
new_df = final_annotations[["post","prefLabel",'Document_Keywords']]
new_df['compulsion'] = 0
new_df['obs-com'] = 0
new_df['obsession'] = 0
```

# Annotate posts with associated classes

```python
#add lables to the dataset based on (compulsion, obs-com and obession) files
for post in compulsion['post']:
    row_number = int(post.split('post')[-1])
    new_df.loc[row_number, 'compulsion'] = 1

for post in obs_com['post']:
    row_number = int(post.split('post')[-1])
    new_df.loc[row_number, 'obs-com'] = 1

for post in obession['post']:
    row_number = int(post.split('post')[-1])
    new_df.loc[row_number, 'obsession'] = 1
```

# extract and save unlabled posts

```python
unlabled_data = new_df[(new_df['compulsion']==0) & (new_df['obs-com']==0) & (new_df['obsession']==0)]
unlabled_data.to_csv('../content/drive/My Drive/post classification/processed_data/unlabled_data.csv',index=False)
```

# save the labled data

```python
labeled_data = new_df[~new_df.index.isin(unlabled_data.index)]
labeled_data.to_csv('../content/drive/My Drive/post classification/processed_data/labled_data.csv',index=False)
```
